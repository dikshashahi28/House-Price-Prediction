```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset  = pd.read_excel("HousePricePrediction_dataset.xlsx")
dataset.head()
```

```python
dataset.shape
```

```python
dtypes = dataset.dtypes
object = dtypes[dtypes == 'object']
print("Categorical varaibles ",": ", len(object))

Float = dtypes[dtypes == 'float']
print("Float Variables : ", len(Float))

Int = dtypes[dtypes == 'int']
print("Integer variales : ", len(Int))

print(list(object.index))

```

```python
numerical_dataset = dataset.select_dtypes(include = ['number'])
plt.figure(figsize=(8, 3))
sns.heatmap(numerical_dataset.corr(), cmap='BrBG', fmt='.2f', linewidths=0.5, annot=True)
plt.show()
```

```python
unique_values = []
for obj in list(object.index):
  unique_values.append(dataset[obj].unique().size)

plt.figure(figsize=(5, 2))
plt.title('No. Unique values of Categorical Features')
plt.xticks(rotation=45)
ax = sns.barplot(x=list(object.index),y=unique_values, palette=['blue', 'red', 'green', 'yellow'])
for bar in ax.containers:
  ax.bar_label(bar, fmt='%.0f')
plt.show()
```
